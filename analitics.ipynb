```python
# Сортировка вставкой
def selection_sort(x):
    for i in range(len(x)):
        swap = i + np.argmin(x[i:])
        (x[i], x[swap]) = (x[swap], x[i])
    return x
```

```python
x = np.array([2, 1, 4, 3, 5])
selection_sort(x)
```

```python
# Случайная сортировка
def bogosort(x):
    while np.any(x[:-1] > x[1:]):
        np.random.shuffle(x)
    return x
```

```python
x = np.array([2, 1, 4, 3, 5])
bogosort(x)
```

```python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                    columns = ['key', 'data1', 'data2'])
df
```

```python
# Агрегирование
df.groupby('key').aggregate(['min', np.median, max])
```

```python
# Еще один удобный паттерн — передача в него словаря,
# связывающего имена столбцов с операциями,
# которые должны быть применены к этим столбцам
df.groupby('key').aggregate({'data1': 'min',
                             'data2': 'max'})
```

```python
# Фильтрация. Операция фильтрации дает возможность опускать данные
# в зависимости от свойств группы
def filter_func(x):
    return x['data2'].std() > 4
print(df);
print(df.groupby('key').std());
print(df.groupby('key').filter(filter_func))
```

```python
# Преобразование. В то время как агрегирующая функция должна возвращать
# сокращенную версию данных, преобразование может вернуть
# версию полного набора данных
df.groupby('key').transform(lambda x: x - x.mean())
```

```python
# Метод apply(). Метод apply() позволяет применять произвольную функцию
# к результатам группировки
def norm_by_data2(x):
    x["data1"] /= x["data2"].sum()
    return x
print(df)
print(df.groupby('key').apply(norm_by_data2))
```

```python
from sklearn.datasets import fetch_lfw_people
faces = fetch_lfw_people(min_faces_per_person=60)
print(faces.target_names)
print(faces.images.shape)
```

```python
from sklearn.svm import SVC
from sklearn.decomposition import PCA as RandomizedPCA
from sklearn.pipeline import make_pipeline
pca = RandomizedPCA(n_components=150, whiten=True, random_state=42)
svc = SVC(kernel='rbf', class_weight='balanced')
model = make_pipeline(pca, svc)
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(faces.data,
                                                faces.target,
                                                random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'svc__C': [1, 5, 10, 50],
              'svc__gamma': [0.0001, 0.0005, 0.001, 0.005]}
grid = GridSearchCV(model, param_grid)
%time grid.fit(Xtrain, ytrain)
print(grid.best_params_)
```

```python
model = grid.best_estimator_
yfit = model.predict(Xtest)
```
